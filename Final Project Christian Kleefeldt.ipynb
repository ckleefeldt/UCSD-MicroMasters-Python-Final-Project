## Matching column names to ensure successful merging

```python
fropen_men.columns = ['FNL1' if x=='FNL.1' else 'FNL2' if x=='FNL.2' else x for x in fropen_men.columns]
fropen_women.columns = ['FNL1' if x=='FNL.1' else 'FNL2' if x=='FNL.2' else x for x in fropen_women.columns]
usopen_women.columns = ['Player1' if x=='Player 1' else 'Player2' if x=='Player 2' else x for x in fropen_women.columns]
usopen_women.columns = ['Round' if x=='ROUND' else x for x in fropen_women.columns]
wimb_men.columns = ['FNL1' if x=='FNL.1' else 'FNL2' if x=='FNL.2' else x for x in wimb_men.columns]
wimb_women.columns = ['FNL1' if x=='FNL.1' else 'FNL2' if x=='FNL.2' else x for x in fropen_men.columns]
wimb_women.columns = ['ST1.1' if x=='ST1.1.1' else 'FNL2' if x=='FNL.2' else x for x in fropen_men.columns]
```

```python
all_men = pd.concat([ausopen_men, fropen_men, usopen_men, wimb_men], sort=False)
all_men.head()
```

```python
all_women = pd.concat([ausopen_women, fropen_women, usopen_women, wimb_women], sort=False)
all_women.head()
```

```python
all_men.describe()
```

```python
all_women.describe()
```

## Dropping all set results from  dataframes since the set results directly indicate the match result and are also a result of the match stats we want to use for analysis. Using a different attribute for results to predict the results would falsify our analysis, we therefore omit them.

## This also solves the issue of missing values if a match ended in 3 or 4 sets instead of 5

## Note: The final number of games won (FNL) will not be dropped. In Tennis, the player who wins the most number of games does not automatically win the match as the number of sets won determines the winner of the match

```python
all_women.drop(['ST1.2', 'ST2.2', 'ST3.2', 'ST1.1', 'ST2.1', 'ST3.1'], axis=1, inplace=True)
all_women.head()
```

```python
all_men.drop(['ST1.2', 'ST2.2', 'ST3.2', 'ST4.2', 'ST5.2', 'ST1.1', 'ST2.1', 'ST3.1', 'ST4.1', 'ST5.1'], axis=1, inplace=True)
all_men.head()
```

## Dropping player names from dataframes

```python
all_men.drop(['Player1', 'Player2'], axis=1, inplace=True)
all_women.drop(['Player1', 'Player2'], axis=1, inplace=True)
```

## Shuffling rows in dataframes to lose the order of tournaments and rounds  

```python
from sklearn.utils import shuffle
all_men_shuffled = shuffle(all_men)
all_women_shuffled = shuffle(all_women)
```
